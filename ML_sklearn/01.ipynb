```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import mglearn
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193] + "\n...")
```

```python
print("Target names: \n{}".format(iris_dataset["target_names"]))
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])
iris_dataframe[:5]
```

```python
# sns.pairplot(data=iris_dataframe)
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', )
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNC
knn = KNC(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print("Test set prediction:\n{}".format(y_pred))
```

```python
print("Test set score:\n{:.2f}".format(knn.score(X_test, y_test)))
```
