```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon", 
                              "Pastrami", "corned beef", "Bacon", "pastrami", "honey ham", "nova lox"], "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {'bacon': 'pig',
                  'pulled pork': 'pig',
                  'pastrami': 'cow',
                  'corned beef': 'cow',
                  'honey ham': 'pig',
                  'nova lox': 'salmon'}
```

```python
lowercased = data['food'].str.lower()
```

```python
lowercased
```

```python
data["animal"] = lowercased.replace(meat_to_animal)
```

```python
data
```

```python
lowercased.map(meat_to_animal)
```
